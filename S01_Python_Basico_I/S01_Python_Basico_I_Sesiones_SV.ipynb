### Ejercicio 3

```py
try:
    nombre = 'Aleph'
    activos = float("2,000")
    pasivos = float("1500") #Aquí puede darse un error por tipos de variables incorrectos
    patrimonio = activos - pasivos 
    razon_end = pasivos/patrimonio #Aqui puede existir una division por cero
    print('El patrimonio de', nombre, "es de", patrimonio, "USD, y su razon de endeudamiento es de", razon_end)
except ZeroDivisionError:
    print("Error: Division por cero")
except TypeError:
    print("Error: Tipos de variables incorrectas")
```
