## Definición y uso de funciones

### Ejercicio 1

Escribe una función llamada `promedios` que calcule los principales promedios de dos números cualquiera y devuelva los resultados a modo de diccionario. Para esto considerar los siguientes estadísticos:

* Media aritmética ($\bar{x}_a$):

$$ \bar{x}_a = \frac{x_1 + x_2}{2} $$

* Media geométrica ($\bar{x}_g$):

$$ \bar{x}_g = \sqrt{|x_1 \times x_2|} $$

* Media armónica ($\bar{x}_h$):

$$ \bar{x}_h = \frac{2}{\frac{1}{x_1} + \frac{1}{x_2}} = \frac{2\times x_1 \times x_2}{x_1 + x_2} $$

Comprueba la función creada con dos números cualquiera.

Si deseas conocer más sobre estos tipos de medias te recomiendo que leas el siguiente link: https://en.wikipedia.org/wiki/Pythagorean_means.
