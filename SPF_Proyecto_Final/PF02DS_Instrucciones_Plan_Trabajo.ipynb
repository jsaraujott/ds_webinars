# Instrucciones Entregable 1 - Plan de trabajo (Documento 2 de 4)

## 1. Propósito

Antes de empezar cualquier proyecto real de análisis o ciencia de datos, es importante planificar las actividades a desarrollar. Por lo tanto, en este entregable se busca que desarrolles un documento claro, lógico y detallado con todos los pasos que te permitan resolver el problema expuesto, mediante el correcto entendimiento del contexto del caso y de los datos que se proporcionan para la ejecución del mismo.

En este sentido, es recomendable que investigues un poco sobre el sector donde opera la empresa del caso, y hagas un dignóstico inicial de los datos que te han sido entregados aplicando técnicas de estadística descriptiva.

## 2. Contenido esperado del entregable

Para que tu entregable sea aprobado, el mismo deberá contener al menos lo siguiente: 

1. **Identificación del problema de negocio:** Indica de forma concreta qué se necesita hacer en el proyecto desde la perspectiva del negocio, y explica porqué realizarlo brindará valor. Para esto, es recomendable que por tu cuenta investigues acerca del sector en el que opera la empresa del caso. 

2. **Descripción de los datos:** Realiza una descripción de los datos con los que trabajarás; en concreto, deberás indicar al menos lo siguiente:
    * El detalle del volumen de tus datos (número de filas y columnas).
    * El nivel de calidad de los datos (existencia de valores perdidos, duplicados, tipos de variables incorrectos, etc.)
    * Un resumen estadístico de las variables existentes (estadísticos de centralidad, dispersión y posición, distribuciones, tablas de frecuencias, etc.).

3. **Objetivo técnico:** Identifica cuáles son las variables de mayor interés en el contexto del caso, específicando concretamente cuál de ellas sería el objetivo de pronóstico de los modelos que construyas, y cuáles serían los mejores atributos a emplear. EStablece también el tipo de solución de aprendizaje computacional a desarrollar.   

3. **Algoritmos de modelamiento:** Plantea y justifica cuáles serían los mejores algoritmos a utilizar de acuerdo a tu criterio, y que permitan dar solución al problema de negocio expuesto. Recuerda que más que la cantidad de algoritmos que pleantees, interesa que los mismos sean capaces de pronosticar adecuadamente la variable objetivo, ante lo cual es recomendable que cumplan con los siguientes requisitos:
    * Deben pertencer al tipo correspondiente en cuanto a la necesidad de clasificar, agrupar o hacer regresión.
    * Debes ser capaz de implementarlos con los datos disponibles.
    * Se deben cumplir los supuestos técnicos básicos en cuanto a atributos y variable objetivo. 
    * Deben ser verificables utilizando criterios rigurosos y robustos de validación técnica.

    Vale señalar que deberás especificar al menos 3 algoritmos distintos, puediendo estos ser los vistos en el curso u otros que investigues por tu propia cuenta.

4. **Métricas de rendimiento:** Define claramente los indicadores que vas a utilizar para comprobar la validez de los algoritmos que implementes. Si ves conveniente, incluso puedes describirlos a modo de fórmulas matemáticas. Junto con cada una de las mátricas, recuerda establecer cuáles serían los umbrales aceptables que deberían alcanzar los resultados del modelamiento, y que satisfacerían los criterios esperables en el contexto del caso. 

5. **Diseño experimental:** Especifica las estrategias que utilizarás en cada uno de los algoritmos para mejorar los resultados base que se alcancen. Recuerda incluir métodos como validación cruzada, ajuste de hhiperparámetros, balanceo de variables, entre otros. Da una breve explicación de cómo estas estrategias serán útiles para alcanzar los umbrales definidos en cada una de las métricas. 

5. **Acciones a ejecutar:** Describe una lista de las tareas que vas a realizar a fin de preparar y ajustar los datos para los algoritmos. Es recomendable que este plan incluya de manera ordenada lo siguiente:
    * Al menos tres actividades vinculadas a la preparación y limpieza de los datos.
    * Al menos tres actividades vinculadas a la ingeniería de datos para los distintos algoritmos (i.e codificación, escalamiento, transformación, particionamiento, etc.).

    Adicionalmente, especifica acciones concretas a desarrollar en lo referente al modelamiento y comunicación de resultados, incluyendo lo siguiente: 

    * Actividades vinculadas al desarrollo de modelos, incluyendo la creación de modelos base, y modelos optimizados mediante las estrategias del diseño experimental definidas.
    * Al menos dos actividades vinculadas a la comunicación de resultados (visualizaciones, informe final, etc.)

    Recuerda que al ser un trabajo de nivel profesional, es importante incluir tiempos estimados para cada una de las actividades. Esta asignación de duraciones no es trivial por lo que toma en cuenta que para **TODO** el proyecto final cuentas con dos semanas y debes ajustarte a esta limitante.

6. **Stakeholders impactados:** Menciona a qué roles o áreas del negocio dentro de la empresa le serán de interés los resultados alcanzados, y explica por qué seleccionaste estos stakeholders en el contexto del caso.

## 3. Aspectos negativos a evitar

Tu entregable debe evitar lo siguiente:

* Copiado literal del enunciado del contexto del caso sin una reformulación que evidencie tu endendimiento del sector, negocio y los datos.
* Formulación de un objetivo técnico genérico, ajeno al problema de negocio o inalcanzable con las herramientas o datos disponibles.
* Enlistado de acciones genéricas que no estén acorde al problema de negocio, los datos o el resto de tu plan de trabajo.
* No referenciar variables objetivo y atributos a utilizar.
* Enfoque exclusivo en aspectos técnicos sin una aclaración respecto a su relación o impacto en el contexto del caso.
* Copiado literal de código generado por IA sin una debida comprensión de tu parte.
* Código que genere errores o interrumpciones inesperadas en la ejecución.  
* Falta de celdas de texto (markdown) que den un mayor contexto y profundidad a tu código.

Vale señalar que si tu entregable presenta alguna de estas características, es probable que el revisor no lo apruebe y debas modificarlo.

## 4. Recomendaciones específicas

* En primer lugar, conviene que sigas este orden para organizar tu trabajo:
    1. Lee con detenimiento el contexto del caso y consulta con tu tutor si tienes alguna duda.
    2. Investiga sobre el sector donde opera la empresa.
    3. Has un breve pero completo diagnóstico de los datos.
    3. Desarrolla el entregable incluyendo todo el contenido solicitado y envíalo para revisión.
    4. Mientras esperas la aprobación de tu entregable, estudia a profundidad sobre los algoritmos que planteaste y la forma adecuada de implementarlos.
    5. Si recibes observaciones por parte del revisor, has los ajustes necesarios en tu documento y reenvíalo.
    6. Una vez aprobado, continua con el siguiente entregable (Código de solución).

* Adicionalmente, te sugerimos que de forma preferente descargues todos los archivos necesarios en tu computadora a fin de que puedas desarrollar tu trabajo de manera local.

* Vinculado a lo anterior, asegúrate de mantener todos los archivos organizados en directorios fáciles de acceder en tu computadora, y has respaldos frecuentes de los documentos más sensibles que generes.
