## 1. Propósito y Objetivos
 
En este proyecto integrarás todos los conocimientos aprendidos a lo largo del programa para resolver problemas realistas de negocio basados en datos y modelos de aprendizaje computacional. En este sentido, se busca que:

* Apliques las habilidades técnicas aprendidas a lo largo del bootcamp (preparación de datos, lenguajes de programación, modelos predictivos, visualización y presentación de información, etc.).
* Apliques igualmente las habilidades estratégicas aprendidas (entendimiento de datos y problemas de negocios, storytelling, pensamiento crítico, desarrollo de conclusiones, etc.)
* Desarrolles entregables concretos que puedan ser entendidos por equipos técnicos y de negocio, cumpliendo con el propósito de solucionar problemas reales.
* Te entrenes en desarrollar un trabajo estructurado, y en comunicar tus hallazgos con claridad y orden a nivel profesional.

## 2. Organización del proyecto

### 2.1 Etapas del proyecto

El proyecto final que desarrollarás se compone de 3 etapas, cada uno con sus propios objetivos y entregables a cumplir: 

1. **Plan de trabajo**: en esta etapa se busca que desarrolles un documento claro, lógico y detallado con todos los pasos que te permitan resolver el problema expuesto en el contexto del caso, mediante el correcto entendimiento del mismo y de los datos que se proporcionan.
2. **Código de solución**: en esta etapa deberás implementar el proceso definido en tu plan de trabajo de tal forma que construyas un modelo predictivo utilizable en el contexto del caso, mediante el debido manejo de tus datos y la aplicación de criterios de experimentación científica y evaluación técnica.
3. **Informe de solución**: en esta etapa tienes que crear un informe que resuma de forma estratégica y pertinente los principales resultados, limitaciones, conclusiones y recomendaciones que has alcanzado en las fases precedentes. 

Cabe señalar que cada uno de los entregables que generes tendrán su propia revisión, y posible reentrega en caso de identificarse observaciones.

### 2.2 Roles y responsabilidades

Para un adecuado desarrollo del proyecto final es importante que tengas claras las responsabilidades que tanto tú, como los demás miembros de TripleTen, tienen:

**Estudiante (Tú):** 

* Actúas como el profesional responsable de la elaboración y entrega de documentos, a modo de un consultor técnico en ciencia de datos.
* Debes proponer, justificar y documentar todas las decisiones que tomes en el contexto del caso de uso.
* Debes mantener todos tus entregables presentables, ordenados, comprensibles y con códigos reproducibles.

**Revisor**

* Aprueba entregables según los criterios definidos en este documento y en cada uno de los documentos específicos de cada etapa. 

**Tutor** 

* Acompaña al estudiante mediante sesiones 1:1 que deberán agendarse en el link correspondiente.

**Success Manager**

## 3. Criterios y recomendaciones generales

### 3.1 Criterios de evaluación generales

A continuación se especifican los criterios generales obligatorios que todos los entregables que realices durante este proyecto deben cumplir:

* **Entendimiento del caso**: alineación con el contexto y el problema de negocio planteado en el caso.
* **Storytelling**: claridad y consistencia en el razonamiento empleado en todos los documentos generados.
* **Habilidad técnica**: Orden y replicabilidad de códigos, y debido manejo, procesamiento, análisis y modelamiento de datos.
* **Habilidad estratégica**: estructuración y comunicación coherente de acciones, implicancias, conclusiones y recomendaciones que den una debida solución al problema de negocio.  

Adicionalmente, en los documentos específicos de cada etapa encontrarás criterios de evaluación propios de cada entregable y que igualmente debes cumplir para que tengas éxito en tu proyecto final.

### 3.2 Recomendaciones generales

Te sugerimos que todos tus entregables cumplan también con lo siguiente:

* Utiliza Jupyter Notebooks (archivos con extensión *.ipynb*) y aprovecha todas sus capacidades en cuanto a celdas de texto (markdowns) y de código. Recuerda que es importante que tus entregables "cuenten una historia", evidencien tus habilidades técnicas y den soluciones concretas al problema de negocio. 
* Usa las herramientas de IA de forma inteligente. Si vas a apoyarte en ellas, no te limites a copiar las recomendaciones de código, sino que entiéndelo y asegúrate que te sea útil en el contexto. Recuerda que al fin y al cabo tú eres responsable de tus entregables y debes ser capaz de explicarlo.
* Valora la simpleza y el pragmatismo. Considera que quienes potencialmente vean tu trabajo son personas sin las mismas habilidades técnicas que tú tienes, pero que deben ser capaces de entender lo que hiciste y las concusiones e implicancias derivadas de tu trabajo.
* Verifica que todas tus conclusiones e implicancias sean lógicas y comprobables, y que aporten un valor agregado en el contexto del caso. Recuerda siempre que una buena conclusión no es una descripción de un resultado alcanzado, sino una implicancia derivada del mismo.
* Tener dudas y necesitar de criterios más experimentados es parte del proceso. No dejes de hacerlo mediante los canales de TripleTen: 
    * Canal de Proyecto de *Discord*
    * Sesiones 1:1 con tu tutor
    * Canal de Co-learning de *Discord*
