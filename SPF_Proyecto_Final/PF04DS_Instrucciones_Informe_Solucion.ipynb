# Instrucciones Entregable 3 - Informe de solución (Documento 4 de 4)

## 1. Propósito

En este último entregable tienes que crear un informe que resuma de forma estratégica y pertinente los principales resultados, limitaciones, conclusiones y recomendaciones que has alcanzado en las fases precedentes de tu proyecto.

Se espera que seas capaz de abstraer la información de mayor relevancia asociada a todo el caso que has estudiado y resuelto, no solamente en cuanto a aspectos técnicos, sino también en aquellos puntos estratégicos del negocio que brindarían potencial valor para la toma de decisiones.

## 2. Contenido esperado del entregable

Para que tu entregable sea aprobado, el mismo debe contener al menos lo siguiente: 

* Descripción ejecutiva (clara y concreta) en la que se especifique el problema de negocio abordado, y la importancia de su resolución y del análisis realizado.
* Descripción ejecutiva de los datos utilizados detallando el nivel de calidad general encontrado y las limitaciones derivadas de esta situación.
* Descripción técnica de la solución planteada donde se especifique lo descrito a continuación:
    * Objetivo técnico.
    * Algoritmos utilizados.
    * Métricas clave y otros indicadores utilizados.
    * Detalle del mejor modelo aplicado especificando el algoritmo que se utilizó, y las calibraciones realizadas al mismo.
    * Resultados de rendimiento técnico alcanzado y su comparación con otros algoritmos probados.
* Estimación del impacto que el modelo tendrá para el negocio en términos de rentabilidad. Esta estimación puede realizarse en términos cualitativos o cuantitativos pero es importante que tenga sentido en el contexto del caso y en cuanto a los resultados alcanzados por el modelamiento.
* Principales conclusiones y recomendaciones alcanzadas que aporten valor para la toma de decisiones de los stakeholders.  

## 3. Aspectos negativos a evitar en el entregable

Tu entregable debe evitar lo siguiente:

* Enfoque exclusivo en aspectos técnicos sin una aclaración respecto a su relación o impacto en el contexto del caso. 
* Interpretaciones equivocadas de inferencias estadísticas o visualizaciones.
* No referenciar aspectos técnicos o desvincularlos de los temas estratégicos relevantes.
* Conclusiones y recomendaciones genéricas que no tengan un aporte significativo en cuanto al contexto estudiado.

Vale señalar que si tu entregable presenta alguna de estas características, es probable que el revisor no lo apruebe y debas modificarlo.

## 4. Recomendaciones específicas

* En primer lugar, conviene que sigas este orden para organizar tu trabajo:
    1. Si bien puedes empezar a escribir una versión borrador de este informe previo a la aprobación del código de la solución por parte del revisor, es recomendable que esperes a la aprobación de este entregable para la inclusión de los aspectos técnicos y las conclusiones.
    2. Una vez que hayas concluido tu documento, envíalo para revisión.
    3. Si recibes observaciones por parte del revisor, has los ajustes necesarios en tu documento y reenvíalo.
    5. Una vez aprobado, ya tienes terminado todo tu Proyecto Final.
    6. Guarda los entregables finales y aprobados en tu portafolio.

* Sigue la estructura metodológica **CAR (Challenge, Actions, Results)** como mecanismo para organizar tu informe de solución:
    * **Contexto**: Describe brevemente el problema de negocio abordado, la importancia del análisis y las hipótesis que se buscaban validar.
    * **Acción**: Explica qué análisis y metodologías se aplicaron, cómo se prepararon los datos y qué enfoque se siguió.
    * **Resultados**: Sintetiza los hallazgos clave y explicar su impacto en el negocio, así como las recomendaciones prácticas derivadas del análisis.

* Si gustas, puedes realizar este entregable como un documento en PDF, presentación de slides, archivo word o quarto. Lo que interesa finalmente es que el nforme pueda ser compartido y abierto por el revisor. Entonces, si utilizas una de estas alternativas incluye un link en el Jupyter Notebook que envíes en la plataforma.

* Debes ser concreto y específico, no repliques todo lo que hayas hecho en tus entregables anteriores, solamente lo más sustancial y que aporte valor en cuanto al contexto del caso.

* Adicionalmente, te sugerimos que de forma preferente descargues todos los archivos necesarios en tu computadora a fin de que puedas desarrollar tu trabajo de manera local.

* Vinculado a lo anterior, asegúrate de mantener todos los archivos organizados en directorios fáciles de acceder en tu computadora, y has respaldos frecuentes de los documentos más sensibles que generes.
