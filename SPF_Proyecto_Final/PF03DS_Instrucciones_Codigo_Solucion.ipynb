# Instrucciones Entregable 2 - Código de solución (Documento 3 de 4)

## 1. Propósito

En este entregable deberás implementar todo lo definido en tu plan de trabajo (entregable 1) de tal forma que construyas un modelo predictivo utilizable en el contexto del caso, mediante el debido manejo de tus datos y la aplicación de criterios de experimentación científica y evaluación técnica.

En este sentido, deberás demostrar tus capacidades de brindar soluciones a problemas reales a partir de un enfoque basado en datos, en el cual desarrolles paso a paso dicha solución, desde los datos originales hasta un modelo reproducible y de alta asertividad.

## 2. Contenido esperado del entregable

Para que tu entregable sea aprobado, el mismo deberá contener al menos lo siguiente: 

1. **Preparación de datos:** Se espera que seas capaz de:
    * Identificar y ajustar los tipos de datos correspondientes con la naturaleza de cada variable.
    * Identificar y dar tratamiento a valores duplicados, perdidos o inconsistencias lógicas derivadas del contexto del caso.
    * Dado que contarás con distintas fuentes de información, deberás vincular las variables relevantes en un solo dataset mediante procesos de concatenación y/o unión.
    * Crear o modificar variables que te aseguren contar con atributos confiables y completos, y una variable objetivo claramente definida.

2. **Ingeniería de datos:** La generalidad de algoritmos de aprendizaje computacional requieren que los datos de los que aprenden cumplan con ciertos criterios técnicos mínimos. Entonces, tu deberás asegurar esto ejecutando las siguientes actividades:
    * Seleccionar atributos pertinentes mediante criterios lógicos, o a través de metodologías de reducción o transformación dimensional. 
    * Codificar atributos no numéricos de tal manera que sean comprensibles por los algoritmos que utilices.
    * Particionar los datos considerando la potencial naturaleza temporal de los mismos y de forma que sea factible una debida evaluación técnica.
    * Escalar o transformar variables de tal forma que se eviten sesgos por un mal dimensionamiento de los datos y/o las magnitudes.
    * Balancear las clases en caso que exista una distribución no uniforme de los casos observados en la variable objetivo.

3. **Modelamiento:** Utilizando los datos limpios y preparados, deberás ahora iniciar con el entrenamiento de los algoritmos que hayas definido en tu plan de trabajo. Con este propósito es importante que incluyas lo siguiente:
    * Un modelo básico cuyos resultados de rendimiento en cuanto a las métricas establecidas en tu plan de trabajo, sirvan como referencia mínima de calidad. En este punto es recomendable la utilización de algoritmos básicos o los clasificadores o regresores *dummy*.
    * Construcción, entrenamiento y validación técnica de los diferentes algoritmos definidos en el plan de trabajo y constrastar su rendimiento en cuanto a los indicadores con el modelo base.
    * Calibración de umbrales y ajuste de hiperparámetros como mecanismo para mejorar los resultados alcanzados.
    * Determinación de la importancia relativa de los atributos en el pronóstico alcanzado del mejor modelo creado a fin de establecer la coherencia conextual del modelo.

4. **Resumen de resultados:** Describe finalmente las principales conclusiones alcanzadas. Recuerda que una buena conclusión no es una descripción de los resultados sino una interpretacion de los mismos en el contexto del caso. En esta etapa igualmente podrías incluir las limitaciones encontradas en cuanto a los datos y los modelos implementados, así como definir potenciales siguientes pasos a realizar.

## 3. Aspectos negativos a evitar

Tu entregable debe evitar lo siguiente:

* Enfoque exclusivo en aspectos técnicos sin una aclaración respecto a su relación o impacto en el contexto del caso. 
* Copiado literal de código generado por IA sin una debida comprensión de tu parte.
* Código que genere errores o interrumpciones inesperadas en la ejecución.  
* Falta de celdas de texto (markdown) que den un mayor contexto y profundidad al código implementado.
* Código que no tenga ninguna relación con la solución o que aporte poco o nada al proceso de preparación, ingeniería y modelamiento planificado.
* Modelos que presenten problemas como sobreajuste o que generen resultados no replicables.  

Vale señalar que si tu entregable presenta alguna de estas características, es probable que el revisor no lo apruebe y debas modificarlo.

## 4. Recomendaciones específicas

* En primer lugar, conviene que sigas este orden para organizar tu trabajo:
    1. Si bien puedes empezar a probar algunos aspectos de tu plan de trabajo previo a que este sea aprobado por el revisor, no es conveniente iniciar con la parte de modelamiento hasta que tu primer entregable sea aprobado.
    2. En este sentido, inicia con tu código de solución definitivo una vez que tu plan de trabajo haya sido aprobado.
    3. Una vez que lo hayas concluido, envíalo para revisión.
    4. Si recibes observaciones por parte del revisor, has los ajustes necesarios en tu documento y reenvíalo.
    5. Una vez aprobado, continua con el siguiente entregable (Informe de solución).

* En vista que tu código debe ser reproducible, te sugerimos que utilices raices aleatorias tipo `random_state`, tanto en las actividades de preparación e ingeniería, como en el modelamiento.

* Adicionalmente, te sugerimos que de forma preferente descargues todos los archivos necesarios en tu computadora a fin de que puedas desarrollar tu trabajo de manera local.

* Vinculado a lo anterior, asegúrate de mantener todos los archivos organizados en directorios fáciles de acceder en tu computadora, y has respaldos frecuentes de los documentos más sensibles que generes.
